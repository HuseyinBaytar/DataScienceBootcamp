# <p style="border-radius:10px; border:#DEB887 solid; padding:25px; background-color: #FFFAF0; font-size:100%;color:#52017A;text-align:center;"> Sampling </p>

### It is a subset within a main population, assumed to represent the characteristics of that main population well.

> Let's say there is a city with an average population of 10,000 people. To find the average age of all of them, we would need to individually approach all 10,000 people and collect their ages to calculate the average. This would require a substantial amount of workforce, so instead, we select a random and unbiased subset of 100 people that represent the 10,000 people well. This allows us to make an inference without having to survey all 10,000 people.

# <p style="border-radius:10px; border:#DEB887 solid; padding:25px; background-color: #FFFAF0; font-size:100%;color:#52017A;text-align:center;"> Hypothesis Testing </p>

### They are statistical methods used to test a belief or proposition. The primary goal in group comparisons is to determine whether any observed differences are due to chance. 

> For example, after making changes to the interface in a mobile application, we want to see if the average daily time users spend on the application has increased. We refer to the time before the interface change as A and after the change as B. We establish a hypothesis that there is no difference in the time spent before and after the interface change, then proceed to test it. Let's say, in the first design, users spent 55 minutes, and in the second design, they spent 58 minutes.

>  Just because the time is 58 minutes, can we say the second design is better? No, we cannot. This is the most critical point of A/B testing. Mathematically, 58 might seem better, but we need to prove whether this difference could have occurred by chance. We need to determine whether this difference is statistically significant

<h3 align="center"><font color='#5E5273'> A/B Test </font></h3>

### We use it to compare the means between two groups.
> H0: The two group means are equal.

> H1: The two group means are not equal.

We will interpret based on the p-value. If the p-value is less than 0.05, we reject H0. The independent two-sample T-test has two assumptions. Normality and Variance Homogeneity, which mean normal distribution and similarity between the two groups.

<center><span style="color:#f7c297;font-family:cursive;font-size:100%"> -Hypotesis- </span></center>
<center><img src="https://i.imgur.com/hpmPgXw.png" width="800" height="800"></center>

> Step 1: Formulate the Hypotheses.

> Step 2: Assumption Check (normality assumption / Variance homogeneity)

> Step 3: Applying the Hypothesis
* If the assumptions are met, apply the independent two-sample T-test, i.e., A/B testing.
* If the assumptions are not met, apply the Mann-Whitney U test.

> Step 4: Interpret the results based on the p-value.
Note:
- If normality is not satisfied, directly go to step 2. If variance homogeneity is not satisfied, go to step 1.
- Prior to the normality check, it might be beneficial to examine and correct for outliers."

```python
df = pd.read_csv("/kaggle/input/tips-dataset/tips.csv")
df.head().T
```

>  Example 1 :Is there a statistically significant difference in the bill averages between smokers and non-smokers?

```python
df.groupby("Smoker").agg({"Total_bill": "mean"})
#It seems like there is a difference, but is it statistically significant? Did this difference occur by chance?
```

```python
#1- Formulate the hypothesis.
#H0: M1=M2
#H1: M1 != M2

#2- Assumption check
#####################Normality assumption
#H0: The normal distribution assumption is satisfied.
#H1: ... is not satisfied.
#We check the normality assumption with the Shapiro-Wilk test
test_stat, pvalue = shapiro(df.loc[df["Smoker"]== "Yes", "Total_bill"])
print('test stat = %.4f, p-value = %.4f' % (test_stat, pvalue))
#Since the p-value is less than 0.05, we reject H0, and the normal distribution assumption is not satisfied.
```

```python
test_stat, pvalue = shapiro(df.loc[df["Smoker"]== "No", "Total_bill"])
print('test stat = %.4f, p-value = %.4f' % (test_stat, pvalue))

#It is 0.00, therefore, H0 is rejected, indicating that the normal distribution assumption is not satisfied.
#We need to use non-parametric tests.
```

```python
##################### Variance homogeneity assumption

#H0: The variances are homogenous.
#H1: The variances are not homogenous.
#Variance homogeneity test
test_stat, pvalue = levene(df.loc[df["Smoker"] == "Yes", "Total_bill"],
df.loc[df["Smoker"]== "No", "Total_bill"])
print('Test stat = %.4f , p-value = %.4f' % (test_stat, pvalue))
#The p-value < 0.05, rejecting H0, the variances are not homogeneous.
```

```python
#3. Applying the hypothesis

#* If the assumptions are met, apply the independent two-sample T-test, i.e., A/B testing (just trying to see, if normality and variance 
#were satisfied, we would use equal_var = True, but since the variance is not satisfied, we will use equal_var = False)
test_stat, pvalue = ttest_ind(df.loc[df["Smoker"] == "Yes", "Total_bill"],df.loc[df["Smoker"]== "No", "Total_bill"],equal_var=True)
print('Test stat = %.4f , p-value = %.4f' % (test_stat, pvalue))

#There is no statistically significant difference between smokers and non-smokers.

#There is no significant difference between their means. H0 cannot be rejected.
```

> Example 2: Is there a statistically significant difference in the average ages between male and female passengers on the Titanic?

```python
df = pd.read_csv("/kaggle/input/titanic/train.csv")
df.head()
```

```python
#it seems there is difference, but is it statistically significant? Did this difference occur by chance?
df.groupby("Sex").agg({"Age": "mean"})
```

```python
#1. Formulate the hypotheses:

#H0: m1 = m2 (There is no significant difference in age averages)
#H1: m1 != m2 (There is a significant difference in age averages)

#2. Examine assumptions
#Normality assumption
#H0: The normal distribution assumption is satisfied.
#H1: ... is not satisfied.

test_stat , pvalue =shapiro(df.loc[df["Sex"] == "female", "Age"].dropna())
print('Test stat = %.4f , p-value = %.4f' % (test_stat, pvalue))
#h0: rejected, normal distribution is not met.
```

```python
test_stat , pvalue =shapiro(df.loc[df["Sex"] == "male", "Age"].dropna())
print('Test stat = %.4f , p-value = %.4f' % (test_stat, pvalue))
#h0: rejected, normal distribution is not met, p-value is less than 0.05.
```

```python
#Variance homogeneity
#H0: Variances are homogeneous.
#H1: Variances are not homogeneous.

test_stat , pvalue =levene(df.loc[df["Sex"] == "female", "Age"].dropna(),df.loc[df["Sex"] == "male", "Age"].dropna())
print('Test stat = %.4f , p-value = %.4f' % (test_stat, pvalue))

#The variances are homogeneous, we cannot reject H0 because the p-value is higher than 0.05. However, since both assumptions are not met,
#we will use non-parametric tests.
```

```python
test_stat , pvalue = mannwhitneyu(df.loc[df["Sex"] == "female", "Age"].dropna(),df.loc[df["Sex"] == "male", "Age"].dropna())

print('Test stat = %.4f , p-value = %.4f' % (test_stat, pvalue))

#h1: There is a statistically significant difference between female and male passengers."
```

> Example3 : Is there a statistically significant difference in ages between individuals with and without diabetes?

```python
df = pd.read_csv("/kaggle/input/diabetes-dataset/diabetes.csv")
df.head()
```

```python
df.groupby('Outcome').agg({'Age': 'mean'})
#1= diabetic, 0= not diabetic
#The mean for 1 is 37, and the mean for 0 is 31. Is this difference due to chance or is it statistically significant?
```

```python
#Step 1: Formulate the Hypotheses.
#h0: m1 = m2 = no difference
#h1: m1 != m2 = there is a difference

#Step 2: Assumption Check (normality assumption / Variance homogeneity)
test_stat, pvalue = shapiro(df.loc[df['Outcome'] == 1, 'Age'].dropna())
print('Test stat = %.4f , p-value = %.4f' % (test_stat, pvalue))
```

```python
test_stat, pvalue = shapiro(df.loc[df['Outcome'] == 0, 'Age'].dropna())
print('Test stat = %.4f , p-value = %.4f' % (test_stat, pvalue))

#Normality assumption is not satisfied, we need to use Mann-Whitney U.
```

```python
#Step 3: Applying the Hypothesis
# * If the assumptions are met, apply the independent two-sample T-test, i.e., A/B testing.
# * If the assumptions are not met, use the Mann-Whitney U test.

test_stat , pvalue = mannwhitneyu(df.loc[df['Outcome'] == 0, 'Age'].dropna(),df.loc[df['Outcome'] == 1, 'Age'].dropna())
print('Test stat = %.4f , p-value = %.4f' % (test_stat, pvalue))

#h0 is rejected, the two means are equal. There is a statistically significant difference."
```

```python
#Two-sample proportion test, comparing two group proportions
#Previously, we had means, now we have proportions

#h0: There is no difference between the conversion rate of the new design and the rate of the old design
#h1: ... there is.
success_count = np.array([300,250])
observation_count = np.array([1000, 1100])

proportions_ztest(count=success_count, nobs=observation_count)

#Since the p-value is less than 0.05, we reject h0, indicating a significant difference between the two rates.
```

> Example 4: Is there a significant difference in survival rates between men and women?

* h0: p1 = p2 -- No difference
* h1: p1 != p2 -- There is a difference

```python
df = pd.read_csv("/kaggle/input/titanic/train.csv")
print(df.loc[df["Sex"] == "female", "Survived"].mean())
print(df.loc[df["Sex"] == "male", ("Survived")].mean())

female_success_count = df.loc[df["Sex"] == "female", "Survived"].sum()
male_success_count = df.loc[df["Sex"] == "male", "Survived"].sum()
```

```python
test_stat, pvalue = proportions_ztest(count=[female_success_count, male_success_count],
nobs=[df.loc[df["Sex"] == "female", "Survived"].shape[0],df.loc[df["Sex"] == "male", "Survived"].shape[0]])
print('Test stat = %.4f , p-value = %.4f' % (test_stat, pvalue))

#h0 is rejected, there is a difference.
```

```python
#Comparison of means for more than two groups
#(ANOVA - Analysis of Variance)
df = pd.read_csv("/kaggle/input/tips-dataset/tips.csv")

#Our hypothesis is: h0: p1=p2=p3 groups are equal

#h1: p1!=p2!=p3 they are not equal

#In terms of the days of the week, is there a difference in the paid bills?
#Weekdays are close to each other, and weekends are close to each other, but there seems to be a difference between weekdays and weekends,
#is it by chance?
df.groupby("Day")["Total_bill"].mean()
```

```python
#h0: m1= m2= m3= m4
#h1: there is a difference
#Assumption check
#If the assumption exists, use one way anova, if the assumption is not met, use kruskal
#h0: The normal distribution assumption is satisfied

for group in list(df["Day"].unique()):
    pvalue = shapiro(df.loc[df["Day"] == group, "Total_bill"])[1]
    print(group, 'p-value: %.4f' % pvalue)

#All are less than 0.05, h0 is rejected, the normality assumption is not met.
```

```python
#h0: The assumption of variance homogeneity is provided.
test_stat, pvalue = levene(df.loc[df["Day"] == "Sun", "Total_bill"],
                           df.loc[df["Day"] == "Sat", "Total_bill"],
                           df.loc[df["Day"] == "Thur", "Total_bill"],
                           df.loc[df["Day"] == "Fri", "Total_bill"])
print('Test stat = %.4f , p-value = %.4f' % (test_stat, pvalue))

#pvalue is 0.5, h0 cannot be rejected, the homogeneity assumption is satisfied, but since it did not pass the normal distribution assumption,
#we will go with non-parametric in any case.
```

```python
#3. Hypothesis testing and p-value interpretation

#parametric anova test: if the assumption were met, which is not the case in this scenario.

f_oneway(df.loc[df["Day"] == "Sun", "Total_bill"],
         df.loc[df["Day"] == "Sat", "Total_bill"],
         df.loc[df["Day"] == "Thur", "Total_bill"],
         df.loc[df["Day"] == "Fri", "Total_bill"])
#p value is less than 0.5, h0 is rejected. There is a statistically significant difference between the groups.
```

```python
#non-parametric test: valid for this scenario
kruskal(df.loc[df["Day"] == "Sun", "Total_bill"],
        df.loc[df["Day"] == "Sat", "Total_bill"],
        df.loc[df["Day"] == "Thur", "Total_bill"],
        df.loc[df["Day"] == "Fri", "Total_bill"])
#As a result of this test, the p-value is less than 0.05, h0 is rejected, there is a statistically significant difference.
```

```python
#We conducted pairwise comparisons to observe how the values of the twos differ.
from statsmodels.stats.multicomp import  MultiComparison
comparison = MultiComparison(df['Total_bill'], df['Day'])
tukey = comparison.tukeyhsd(0.05)
print(tukey.summary())
```
